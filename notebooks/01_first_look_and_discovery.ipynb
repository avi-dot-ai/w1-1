#2. Discover a real truth yourself


```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

#Discovery A: for pages with impressions_90d > 0

```python
filtered_df = df[df["impressions_90d"] > 0]
corr = filtered_df["search_volume"].corr(filtered_df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d (for pages with impressions > 0): {corr:.3f}")

if abs(corr) < 0.1:
    print("Observation: Even for pages with some impressions, the correlation remains very low, suggesting search volume is still a weak predictor of actual traffic.")
elif corr > 0.1:
    print("Observation: The correlation increased slightly, but still indicates a weak relationship between search volume and traffic.")
else:
    print("Observation: The correlation became slightly more negative, but still indicates a weak relationship between search volume and traffic.")
```

### Discovery B: Content Type with Worst CTR at the Same Position Tier

```python
visible = df[df["impressions_90d"] >= 100]

# Group by position_tier and content_type, then calculate the median CTR
median_ctr_by_type_pos = visible.groupby(["position_tier", "content_type"])[["ctr"]].median().reset_index()

# Sort by position_tier and then by CTR to easily identify the worst performing content types
median_ctr_by_type_pos_sorted = median_ctr_by_type_pos.sort_values(by=["position_tier", "ctr"], ascending=[True, True])

print("Median CTR by Position Tier and Content Type (sorted by CTR within each tier):\n")
display(median_ctr_by_type_pos_sorted.round(4))

print("\nObservation: For each position tier, the 'guide' content type often exhibits a lower median CTR compared to others, suggesting it might be less effective at converting impressions into clicks, especially in lower position tiers.")
```

### Discovery C: Content Age vs. Trend Direction

```python
# Group by trend_direction and calculate the median content_age_days
median_age_by_trend = df.groupby("trend_direction")["content_age_days"].median().reset_index()

print("Median Content Age by Trend Direction:\n")
display(median_age_by_trend.round(0))

print("\nObservation: Content classified as 'new' has a significantly lower median age, as expected. Beyond that, the median age for 'down', 'flat', 'stable', and 'up' trends are relatively similar, suggesting that for established content, age alone is not a primary driver of trend direction in this dataset.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar plot of median content age by trend direction
fig = plt.figure(figsize=(10, 6))
sns.barplot(x='trend_direction', y='content_age_days', data=median_age_by_trend, palette='viridis')
plt.title('Median Content Age by Trend Direction')
plt.xlabel('Trend Direction')
plt.ylabel('Median Content Age (Days)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Discovery D: Average Position vs. CTR

```python
corr_avg_pos_ctr = df['avg_position'].corr(df['ctr'])
print(f"Correlation between avg_position and ctr: {corr_avg_pos_ctr:.3f}")

if corr_avg_pos_ctr < -0.5:
    print("Observation: There is a strong negative correlation, indicating that as average position improves (becomes a smaller number), CTR significantly increases. This is a crucial signal for ranking.")
elif corr_avg_pos_ctr < -0.2:
    print("Observation: There is a moderate negative correlation, suggesting that better average positions tend to lead to higher CTRs.")
elif abs(corr_avg_pos_ctr) < 0.2:
    print("Observation: The correlation is very low, implying that average position has little direct linear relationship with CTR in this dataset. Other factors might be more influential.")
else:
    print("Observation: The correlation is weak, but there is some relationship between average position and CTR.")
```
