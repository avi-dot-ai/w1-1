# **1. My lane (or freestyle) and why**

**Lane: Ranking Signal Analysis.**
I chose the Ranking Signal Analysis lane because During the initial exploration, I found that search volume has almost no linear relationship with impressions (correlation ≈ 0.001), and average position also showed only a weak linear relationship with CTR (correlation ≈ -0.073). These observations suggest that ranking performance is influenced by multiple interacting factors rather than a single metric. Therefore, I chose the Ranking Signal Analysis lane to investigate which combinations of page-level signals are most useful for understanding and improving search performance. And understanding the factors that influence search rankings has broad applications in search engines, recommendation systems, and AI-driven products. My goal is to identify which measurable signals are most strongly associated with higher-ranking pages and translate those findings into evidence-backed recommendations. This project combines exploratory data analysis, statistical reasoning, feature engineering, and machine learning, making it a strong opportunity to develop practical ML skills while solving a real business problem.

# **2. The question: decision, action, cost of a wrong call**

**Research Question:**

Which page-level signals are most strongly associated with higher search rankings, and which of these should be prioritized to improve organic search performance?

**Decision:**

Determine which ranking signals deserve the highest priority for optimization.

**Action:**

SEO specialists or content teams can optimize pages by improving the most influential signals identified through the analysis.

**Unit of Analysis:**

One webpage (one URL).

**Output:**

A ranked report of important ranking signals with supporting evidence, feature importance analysis, and actionable recommendations.

**Cost of a Wrong Recommendation:**

If unimportant signals are prioritized, teams may spend significant engineering and content resources on changes that do not improve rankings, resulting in wasted time, higher costs, and missed opportunities to improve search visibility.

**Why ML Helps:**

Ranking performance depends on many interacting variables that are difficult to evaluate manually. Data analysis and machine learning can identify meaningful patterns, quantify relationships between features and rankings, and prioritize the factors most likely to influence performance. The objective is to support better decisions rather than replace human judgment.
