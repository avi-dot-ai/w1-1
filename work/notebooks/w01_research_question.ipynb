# **1. My lane (or freestyle) and why**

**Lane: Ranking Signal Analysis.**
I chose the Ranking Signal Analysis lane because my initial exploratory analysis showed that search performance cannot be explained by a single metric alone. While some commonly assumed signals (such as search volume) showed very weak relationships with traffic, initial analyses indicated that impressions and average position consistently emerged as the strongest predictors in simple decision-tree models. This motivated me to investigate which ranking signals are most informative, how they interact, and how they can be translated into evidence-backed recommendations for improving search performance.

# **2. The question: decision, action, cost of a wrong call**

**Research Question:**

Which page-level signals are most strongly associated with higher search rankings, and which of these should be prioritized to improve organic search performance?

**Decision:**

Determine which ranking signals deserve the highest priority for optimization.

**Action:**

SEO specialists or content teams can optimize pages by improving the most influential signals identified through the analysis.

**Unit of Analysis:**

One webpage (one URL).

**Output:**

A ranked report of important ranking signals with supporting evidence, feature importance analysis, and actionable recommendations.

**Cost of a Wrong Recommendation:**

If unimportant signals are prioritized, teams may spend significant engineering and content resources on changes that do not improve rankings, resulting in wasted time, higher costs, and missed opportunities to improve search visibility.

**Why ML Helps:**

Search performance depends on multiple interacting signals, making it difficult to derive reliable decision rules manually. Machine learning can identify influential features, evaluate interpretable ranking rules, and estimate how well those rules generalize to unseen clients. The goal is to support decision-making with evidence rather than maximize predictive accuracy alone.
